5. Se busca en las columnas que tienen horario, el valor de "0" ó "0.0" y saber si existe esta hora en alguna columna, de lo contrario se asigna el valor de 0 para los registros que esten nulos

```python
busqueda = [0.0, 0]

# Se verifica si hay registros con el valor que se cambiara los 'nan' el cual sera '0.0'
df1 = f_2015[f_2015['Sched_Dep_Time'].isin(busqueda)]
df2 = f_2015[f_2015['Dep_Time'].isin(busqueda)]
df3 = f_2015[f_2015['Sched_Arr_Time'].isin(busqueda)]
df4 = f_2015[f_2015['Arr_Time'].isin(busqueda)]
df5 = f_2015[f_2015['Wheels_Off'].isin(busqueda)]
df6 = f_2015[f_2015['Wheels_On'].isin(busqueda)]


# Se muestra si hay algun registro con los valores '0.0' ó '0'
print('\n-------- Sched_Dep_Time --------\n')
print(len(df1['Sched_Dep_Time']))
print('\n-------- Dep_Time --------------\n')
print(len(df2['Dep_Time']))
print('\n-------- Sched_Arr_Time --------\n')
print(len(df3['Sched_Arr_Time']))
print('\n-------- Arr_Time --------------\n')
print(len(df4['Arr_Time']))
print('\n-------- Wheels_Off ------------\n')
print(len(df5['Wheels_Off']))
print('\n-------- Wheels_On -------------\n')
print(len(df6['Wheels_On']))
```

```python
# Se remplazan los valores nulos por 0
# f_2015['Sched_Dep_Time'].fillna(0, inplace = True)    # No es necesario usar
f_2015['Dep_Time'].fillna(0, inplace = True)
# f_2015['Sched_Arr_Time'].fillna(0, inplace = True)    # No es necesario usar
f_2015['Arr_Time'].fillna(0, inplace = True)
f_2015['Wheels_Off'].fillna(0, inplace = True)
f_2015['Wheels_On'].fillna(0, inplace = True)

# Se transforman los datos para que sea mas legible y se guarda como cadena de texto con el formato correcto de hora
f_2015['Sched_Dep_Time'] = pd.to_datetime(f_2015['Sched_Dep_Time'].astype(str).str.zfill(4), format='%H%M').dt.strftime('%H:%M')
f_2015['Dep_Time'] = pd.to_datetime(f_2015['Dep_Time'].astype(int).astype(str).str.zfill(4), format='%H%M', errors =  'coerce').dt.strftime('%H:%M')
f_2015['Arr_Time'] = pd.to_datetime(f_2015['Arr_Time'].astype(int).astype(str).str.zfill(4), format='%H%M', errors =  'coerce').dt.strftime('%H:%M')
f_2015['Wheels_Off'] = pd.to_datetime(f_2015['Wheels_Off'].astype(int).astype(str).str.zfill(4), format='%H%M', errors =  'coerce').dt.strftime('%H:%M')
f_2015['Wheels_On'] = pd.to_datetime(f_2015['Wheels_On'].astype(int).astype(str).str.zfill(4), format='%H%M', errors =  'coerce').dt.strftime('%H:%M')
f_2015['Sched_Arr_Time'] = pd.to_datetime(f_2015['Sched_Arr_Time'].astype(str).str.zfill(4), format='%H%M', errors = 'coerce').dt.strftime('%H:%M')
```

- ### Diccionario <br>

- <b>Year:</b> Año del vuelo

- <b>Month:</b> Mes del vuelo (1 para enero, 2 para febrero, etc.).

- <b>Day_Of_Month:</b> Describe el dia del vuelo entre 1 y 31 

- <b>Day_Of_Week:</b> Día de la semana del vuelo (1 para lunes, 2 para martes, etc.).  

- <b>Airline_Code:</b> Codigo de la aerolinea 

- <b>Flight_Num:</b> Numero de vuelo

- <b>Tail_Num:</b> Número de cola de la aeronave.

- <b>Origin_Airport:</b> Código del aeropuerto de origen.

- <b>Dest_Airport:</b> Código del aeropuerto de destino.

- <b>Sched_Dep_Time:</b> Hora programada de salida.

- <b>Dep_Time:</b> Hora real de salida.

- <b>Dep_Delay:</b> Retraso en la salida (en minutos, positivo si hay retraso).

- <b>Sched_Arr_Time:</b> Hora programada de llegada. 

- <b>Arr_Time:</b> Hora real de llegada.

- <b>Arr_Delay:</b> Retraso en la llegada (en minutos, positivo si hay retraso).

- <b>Taxi_Out:</b> Tiempo de rodaje desde la puerta de embarque hasta la pista de despegue en minutos.

- <b>Wheels_Off:</b> Hora en la que las ruedas de la aeronave dejan el suelo.

- <b>Wheels_On:</b> Hora en la que las ruedas de la aeronave tocan el suelo.

- <b>Taxi_In:</b> Tiempo de rodaje desde la pista de aterrizaje hasta la puerta de llegada en minutos.

- <b>Sched_Elapsed_Time:</b> Tiempo de vuelo programado en minutos.      

- <b>Elapsed_Time:</b> Tiempo real transcurrido durante el vuelo en minutos.   

- <b>Air_Time:</b> Tiempo real en el aire en minutos.

- <b>Distance:</b> Distancia del vuelo en millas.

- <b>Diverted:</b> Desviasion del vuelo, 1 (SI) - 0 (NO).

- <b>Cancelled:</b> Indicador de si el vuelo fue cancelado. 1 (SI) - 0 (NO)

- <b>Cancellation_Reason:</b> Motivo de la cancelacion (A = transportista, B = clima C = NAS, D = seguridad)                

- <b>Nas_Delay:</b> Tiempo de retraso debido al sistema de control del tráfico aéreo.

- <b>Security_Delay:</b> Tiempo de retraso debido a problemas de seguridad.   

- <b>Airline_Delay:</b> Tiempo de retraso atribuido a la aerolínea.  

- <b>Late_Aircraft_Delay:</b> Tiempo de retraso atribuido a la llegada tardía de la aeronave anterior.         

- <b>Weather_Delay:</b> Tiempo de retraso debido a condiciones climáticas.

2. Se remplaza el valor Nan por la fecha de '01/01/1900' en la columna 'issue_date', este cambio solo se hace el la fila 1745 ya que es la unica que tiene el valor Nan
